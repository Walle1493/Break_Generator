绘制生成指标

```python
import numpy as np
import matplotlib.pyplot as plt
bart = np.array([0.494, 0.742, 0.530])
t5 = np.array([0.490, 0.732, 0.534])
x = np.arange(len(bart))
metrics = ["BLEU", "METEOR", "ROUGE"]

plt.figure(figsize=(8,6), dpi=80)
plt.bar(x-0.2, bart, width=0.36, label="Bart-Large", color="#8CA9D3")
plt.bar(x+0.2, t5, width=0.36, label="T5-Large", color="#F26659")

for i, j in zip(x, bart):
    plt.text(i-0.3, j+0.01, bart[i])
for i, j in zip(x, t5):
    plt.text(i+0.1, j+0.01, t5[i])

plt.xticks(x, metrics)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# bart_data
bart_hop2 = np.array([0.534, 0.777, 0.560])
bart_hop3 = np.array([0.473, 0.728, 0.516])
bart_hop4 = np.array([0.407, 0.661, 0.463])
# t5_data
t5_hop2 = np.array([0.560, 0.785, 0.585])
t5_hop3 = np.array([0.442, 0.704, 0.499])
t5_hop4 = np.array([0.362, 0.619, 0.442])

x = np.arange(len(bart_hop2))
metrics = ["BLEU", "METEOR", "ROUGE"]

# bart bar
plt.figure(figsize=(16,6), dpi=80)
plt.subplot(121)
plt.bar(x-0.27, bart_hop2, width=0.21, label="Bart (2-hop)", color="#EBC999")
plt.bar(x, bart_hop3, width=0.21, label="Bart (3-hop)", color="#8CA9D3")
plt.bar(x+0.27, bart_hop4, width=0.21, label="Bart (4-hop)", color="#F26659")

for i, j in zip(x, bart_hop2):
    plt.text(i-0.38, j+0.01, bart_hop2[i])
for i, j in zip(x, bart_hop3):
    plt.text(i-0.11, j+0.01, bart_hop3[i])
for i, j in zip(x, bart_hop4):
    plt.text(i+0.16, j+0.01, bart_hop4[i])

plt.xticks(x, metrics)
plt.legend()

# bart bar
plt.subplot(122)
plt.bar(x-0.27, t5_hop2, width=0.21, label="T5 (2-hop)", color="#EBC999")
plt.bar(x, t5_hop3, width=0.21, label="T5 (3-hop)", color="#8CA9D3")
plt.bar(x+0.27, t5_hop4, width=0.21, label="T5 (4-hop)", color="#F26659")

for i, j in zip(x, t5_hop2):
    plt.text(i-0.38, j+0.01, t5_hop2[i])
for i, j in zip(x, t5_hop3):
    plt.text(i-0.11, j+0.01, t5_hop3[i])
for i, j in zip(x, t5_hop4):
    plt.text(i+0.16, j+0.01, t5_hop4[i])

plt.xticks(x, metrics)
plt.legend()
plt.show()
```
