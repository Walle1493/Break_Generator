绘制生成指标

```python
import numpy as np
import matplotlib.pyplot as plt
bart = np.array([0.494, 0.742, 0.530])
t5 = np.array([0.490, 0.732, 0.534])
x = np.arange(len(bart))
metrics = ["BLEU", "METEOR", "ROUGE"]

plt.figure(figsize=(8,6), dpi=80)
plt.bar(x-0.2, bart, width=0.36, label="Bart-Large", color="#8CA9D3")
plt.bar(x+0.2, t5, width=0.36, label="T5-Large", color="#F26659")

for i, j in zip(x, bart):
    plt.text(i-0.3, j+0.01, bart[i])
for i, j in zip(x, t5):
    plt.text(i+0.1, j+0.01, t5[i])

plt.xticks(x, metrics)
plt.legend()
plt.show()
```

比较seq2seq和multi-step的差异

```python
import numpy as np
import matplotlib.pyplot as plt
seq2seq = np.array([45.1, 74.9, 49.4])
multi_step = np.array([48.9, 71.9, 55.5])
x = np.arange(len(seq2seq))
metrics = ["BLEU", "METEOR", "ROUGE"]

plt.figure(figsize=(8,6), dpi=80)
plt.bar(x-0.2, seq2seq, width=0.36, label="seq2seq", color="#8CA9D3")
plt.bar(x+0.2, multi_step, width=0.36, label="multi-step", color="#F26659")

for i, j in zip(x, seq2seq):
    plt.text(i-0.3, j+1, seq2seq[i])
for i, j in zip(x, multi_step):
    plt.text(i+0.1, j+1, multi_step[i])

plt.xticks(x, metrics)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# bart_data
seq_hop2 = np.array([49.2, 78.8, 53.3])
seq_hop3 = np.array([41.8, 72.1, 46.3])
seq_hop4 = np.array([38.8, 68.0, 43.1])
# t5_data
multi_hop2 = np.array([52.2, 73.9, 58.6])
multi_hop3 = np.array([46.1, 70.9, 52.6])
multi_hop4 = np.array([43.5, 67.5, 51.2])

x = np.arange(len(seq_hop2))
metrics = ["BLEU", "METEOR", "ROUGE"]

# seq2seq bar
plt.figure(figsize=(16,6), dpi=80)
plt.subplot(121)
plt.bar(x-0.27, seq_hop2, width=0.21, label="seq2seq (2-hop)", color="#EBC999")
plt.bar(x, seq_hop3, width=0.21, label="seq2seq (3-hop)", color="#8CA9D3")
plt.bar(x+0.27, seq_hop4, width=0.21, label="seq2seq (4-hop)", color="#F26659")

for i, j in zip(x, seq_hop2):
    plt.text(i-0.36, j+1, seq_hop2[i])
for i, j in zip(x, seq_hop3):
    plt.text(i-0.09, j+1, seq_hop3[i])
for i, j in zip(x, seq_hop4):
    plt.text(i+0.18, j+1, seq_hop4[i])

plt.xticks(x, metrics)
plt.legend()

# multi-step bar
plt.subplot(122)
plt.bar(x-0.27, multi_hop2, width=0.21, label="multi-step (2-hop)", color="#EBC999")
plt.bar(x, multi_hop3, width=0.21, label="multi-step (3-hop)", color="#8CA9D3")
plt.bar(x+0.27, multi_hop4, width=0.21, label="multi-step (4-hop)", color="#F26659")

for i, j in zip(x, multi_hop2):
    plt.text(i-0.36, j+1, multi_hop2[i])
for i, j in zip(x, multi_hop3):
    plt.text(i-0.09, j+1, multi_hop3[i])
for i, j in zip(x, multi_hop4):
    plt.text(i+0.18, j+1, multi_hop4[i])

plt.xticks(x, metrics)
plt.legend()
plt.show()
```
