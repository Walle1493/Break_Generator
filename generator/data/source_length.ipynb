```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8), dpi=100)

# Doc
train_source = np.load("train.source.npy")
train_target = np.load("train.target.npy")
dev_source = np.load("dev.source.npy")
dev_target = np.load("dev.target.npy")

plt.subplot(221)
n1, bins1, _ = plt.hist(train_source, color="#8CA9D3", label="train_source", bins=10, range=(0, 120), rwidth=0.9, edgecolor="#888")
plt.xlabel("length")
plt.legend()

plt.subplot(222)
n2, bins2, _ = plt.hist(train_target, color="#F26659", label="train_target", bins=7, range=(0, 35), rwidth=0.9, edgecolor="#888")
plt.xlabel("length")
plt.legend()

plt.subplot(223)
n1, bins1, _ = plt.hist(dev_source, color="#8CA9D3", label="dev_source", bins=10, range=(0, 120), rwidth=0.9, edgecolor="#888")
plt.xlabel("length")
plt.legend()

plt.subplot(224)
n2, bins2, _ = plt.hist(dev_target, color="#F26659", label="dev_target", bins=7, range=(0, 35), rwidth=0.9, edgecolor="#888")
plt.xlabel("length")
plt.legend()

# for i in range(len(n1)):
#     plt.text(bins1[i]+1.5, n1[i]+300, int(n1[i]))
#     if i in (1,2,3):
#         plt.text(bins2[i]+1.5, n2[i]+150, int(n2[i]))

plt.show()
```
